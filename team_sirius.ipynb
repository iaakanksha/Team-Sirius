# Data Preprocessing

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le_yummy = LabelEncoder()
le_yummy.fit(['Yes','No'])
data.iloc[:,0] = le_yummy.transform(data.iloc[:,0])

le_cheap = LabelEncoder()
le_cheap.fit(['Yes','No'])
data.iloc[:,6] = le_cheap.transform(data.iloc[:,6])

le_fast = LabelEncoder()
le_fast.fit(['Yes','No'])
data.iloc[:,5] = le_fast.transform(data.iloc[:,5])

le_greasy = LabelEncoder()
le_greasy.fit(['Yes','No'])
data.iloc[:,4] = le_greasy.transform(data.iloc[:,4])

le_spicy = LabelEncoder()
le_spicy.fit(['Yes','No'])
data.iloc[:,2] = le_spicy.transform(data.iloc[:,2])

le_fattening = LabelEncoder()
le_fattening.fit(['Yes','No'])
data.iloc[:,3] = le_fattening.transform(data.iloc[:,3])

le_convenient = LabelEncoder()
le_convenient.fit(['Yes','No'])
data.iloc[:,1] = le_convenient.transform(data.iloc[:,1])

le_Like = LabelEncoder()
le_Like.fit(['-3', '+2', '+1', '+4', 'I love it!+5', 'I hate it!-5', '-2', '+3',
       '0', '-4', '-1'])
data.iloc[:,11] = le_Like.transform(data.iloc[:,11])

le_disgusting = LabelEncoder()
le_disgusting.fit(['Yes','No'])
data.iloc[:,10] = le_disgusting.transform(data.iloc[:,10])

le_Gender = LabelEncoder()
le_Gender.fit(['Male','Female'])
data.iloc[:,14] = le_Gender.transform(data.iloc[:,14])

le_healthy = LabelEncoder()
le_healthy.fit(['Yes','No'])
data.iloc[:,9] = le_healthy.transform(data.iloc[:,9])

le_expensive = LabelEncoder()
le_expensive.fit(['Yes','No'])
data.iloc[:,8] = le_expensive.transform(data.iloc[:,8])

le_tasty = LabelEncoder()
le_tasty.fit(['Yes','No'])
data.iloc[:,7] = le_tasty.transform(data.iloc[:,7])

le_VisitFrequency = LabelEncoder()
le_VisitFrequency.fit(['Every three months','Once a week','Once a month','Once a year','Every three months','More than once a week', 'Never'])
data.iloc[:,13] = le_VisitFrequency.transform(data.iloc[:,13])
```

# Principal Component Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
pca.fit(data)

print('Eignevalues:')
print(pca.explained_variance_)
print()

print('Variance:',pca.explained_variance_ratio_*100)

```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_*100))
plt.xlabel('No. of components')
plt.ylabel('Variance')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
prinComp = pca.fit_transform(data)
pca_df = pd.DataFrame(data = prinComp, columns = ['principal component 1','principal component 2'])

sns.heatmap(pca_df.corr(), annot=True)
```

# KMeans

```python
from sklearn.cluster import KMeans
cluster_sumofsquare = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_df)
    cluster_sumofsquare.append(kmeans.inertia_)

plt.figure(figsize=(6,6))    
plt.plot(cluster_sumofsquare)
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
model = (KMeans(n_clusters = 5) )

model.fit(pca_df)
model
```

```python
labels2 = model.labels_
centroids2 = model.cluster_centers_

print(labels2)  
print(centroids2)
```

```python
pca_df['label'] = labels2 

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df["principal component 1"][pca_df.label == 0], pca_df["principal component 2"][pca_df.label == 0], c='blue', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 1], pca_df["principal component 2"][pca_df.label == 1], c='red', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 2], pca_df["principal component 2"][pca_df.label == 2], c='green', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 3], pca_df["principal component 2"][pca_df.label == 3], c='orange', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 4], pca_df["principal component 2"][pca_df.label == 4], c='purple', s=60)
ax.view_init(30, 210)
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")

plt.show()
```

```python
km_5 = KMeans(n_clusters=5, init='k-means++', random_state=0)
km_5.fit(pca_df)
centroids = km_5.cluster_centers_
pca_df['Labels'] = km_5.labels_


plt.figure(figsize=(10, 6))

sns.scatterplot(pca_df['principal component 1'], pca_df['principal component 2'], hue=pca_df['Labels'], 
                palette=sns.color_palette('hls', 5))

plt.scatter(centroids[:,0], centroids[:,1], c='red',s=300)

plt.title('K-Means Clustering',fontsize=18)
plt.show()
```
